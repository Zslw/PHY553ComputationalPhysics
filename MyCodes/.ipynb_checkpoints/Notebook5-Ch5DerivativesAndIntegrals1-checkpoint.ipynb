# 5.2.1 Second Derivatives

## Exercise - Landau Section 5.2 (at end): Compare derivative algorithms (box counting)

### 5.2.1.1 Assessment

The programming for numerical differentiation is simple:

``` pseudocode
FD = ( y(t+h)− y(t) ) /h;                                       // Forward diff
CD = ( y(t+h/2)− y(t−h/2) ) /h;                                 // Central diff
ED = (8∗ (y(t+h/4)−y(t−h/4))− (y(t+h/2)−y(t−h/2)))/3/h;         //extra
```

1) Use forward-, central-, and extrapolated-difference algorithms to differentiate the functions $cos(t)$ and $e^t$ at $t= 0.1 \text{,} 1.0$ , and $100$.

- a) Print out the derivative and its relative error $\boldsymbol{\varepsilon}$ as functions of $h$. Reduce the step size $h$ until the error equals machine precision $h\simeq \epsilon_m$.

- b) Plot $log_{10}|\boldsymbol{\varepsilon}|$ versus $log_{10}h$ and check whether the number of decimal places obtained agrees with the estimates in the text.

- c) See if you can identify regions where algorithmic (series truncation) error dominates at large $h$ and round-off error at small $h$ in your plot. Do the slopes agree with our model's predictions?

````python
import numpy as np

def selectFunction(f): 
    '''
    This function allow to select between the 2 functions 
    Cos(x) or e^x for this exercise
    '''
    functions = {
        'cos': (np.cos, lambda x: -np.sin(x)),  # cos(x) and its derivative -sin(x)
        'exp': (np.exp, np.exp)  # e^x and its derivative e^x
    }
    
    return functions.get(f.lower(), (None, None))  # Returns (None, None) if invalid


def forwardDiff(f,x,h): #( y(t+h)− y(t) ) /h
    '''
    This function performs a forward differentiation
    It inputs a function f, position x, size steps h
    Returns the numerical value of the derivative at that point x

    example:
    forwardDiff(Cos,0.1,10**(-4))) =  -0.8414979985205129
    '''
    func, _ = selectFunction(f)
    return (func(x + h) - func(x)) / h

def centralDiff(f,x,h): #( y(t+h/2)− y(t−h/2) ) /h
    '''
    This function performs a central differentiation
    It inputs a function f, position x, size steps h
    Returns the numerical value of the derivative at that point x

    example:
    forwardDiff(Cos,0.1,10**(-4))) =  -0.8414979985205129
    '''
    func, _ = selectFunction(f)
    return (func(x + h/2) - func(x - h/2)) / h

def extraDiff(f,x,h): # (8∗ (y(t+h/4)−y(t−h/4))− (y(t+h/2)−y(t−h/2)))/3/h
    '''
    This function performs a extrapolating differentiation
    It inputs a function f, position x, size steps h
    Returns the numerical value of the derivative at that point x

    example:
    forwardDiff(Cos,0.1,10**(-4))) =  -0.8414979985205129
    '''
    func, _ = selectFunction(f)
    return (8 * (func(x + h/4) - func(x - h/4)) - (func(x + h/2) - func(x - h/2))) / (3 * h)
````

````python
print(forwardDiff('cos',1,10**(-4))) #make h really small h=10**(-4)
print(centralDiff('cos',1,10**(-4)))
print(extraDiff('cos',1,10**(-4)))
print(forwardDiff('exp',1,10**(-4)))
print(centralDiff('exp',1,10**(-4)))
print(extraDiff('exp',1,10**(-4)))
````

1) a) we can iterate these previous functions to calculate the values 
that we want and compare with the exact thing, 

I realized I can adapt the selectFunction to return a tuple with the exact 
derivative and use it in this one as well

````python
def compute_errors(f, x_values, h_values):
    '''
    Computes numerical derivatives and relative errors
    '''
    results = {} #lets use a dictionary this time, so we can actually store everything here

    for x in x_values:
        func, true_derivative = selectFunction(f) # use my selectFunction here
        true_val = true_derivative(x) #store the true values of the derivative
        errors = []

        for h in h_values: #computate each derivative with its respective method and its error
            fwd = forwardDiff(f, x, h)
            cen = centralDiff(f, x, h)
            ext = extraDiff(f, x, h)

            err_fwd = abs((fwd - true_val) / true_val)
            err_cen = abs((cen - true_val) / true_val)
            err_ext = abs((ext - true_val) / true_val)

            errors.append((h, fwd, err_fwd, cen, err_cen, ext, err_ext))

        results[x] = errors

    return results
````

Let's run the previous function to see results

````python
h_values = np.logspace(-16, 0, 50)  # Step sizes from 10^-16 to 1
x_values = [0.1, 1.0, 100]  # Given points
results = compute_errors('cos', x_values, h_values)
len(results)
````

So now results is size 3, that is each x value, with a lot of values for each h inside
we can try printing `results` from here but instead, lets do it again with both functions and

````python
h_values = np.logspace(-16, 0, 50)  # Step sizes from 10^-16 to 1
x_values = [0.1, 1.0, 100]  # Given points

for f in ['cos', 'exp']:
    print(f"\nFunction: {f.upper()}")
    results = compute_errors(f, x_values, h_values)

    for x, errors in results.items():
        print(f"\nAt x = {x}:")
        print(f"{'h':<15} {'Forward':<15} {'Err_Fwd':<15} {'Central':<15} {'Err_Cen':<15} {'Extrapolated':<15} {'Err_Extra':<15}")
        for h, fwd, err_fwd, cen, err_cen, ext, err_ext in errors:
            print(f"{h:<15.1e} {fwd:<15.5e} {err_fwd:<15.5e} {cen:<15.5e} {err_cen:<15.5e} {ext:<15.5e} {err_ext:<15.5e}")
````

I haven't decreased h until machine precision so lets try that now

````python
# Parameters
epsilon_m = np.finfo(float).eps  # Machine precision found with google, im not doing that cycle from Ch2 again

for x in x_values:
    print(f"\nEvaluating at x = {x}")
    print(f"{'h':<12} {'Forward Error':<20} {'Central Error':<20} {'Extrapolated Error':<20}") #header
    #:<12 left alignment 12 characters
    
    h = 1e-1  # Initial step size

    while h > epsilon_m:
        # Compute numerical derivatives
        fwd = forwardDiff('cos', x, h)
        cent = centralDiff('cos', x, h)
        extra = extraDiff('cos', x, h)

        # Get exact derivative
        _, dfunc = selectFunction('cos')
        true_derivative = dfunc(x)

        # Compute relative errors
        fwd_error = abs(fwd - true_derivative) / abs(true_derivative)
        cent_error = abs(cent - true_derivative) / abs(true_derivative)
        extra_error = abs(extra - true_derivative) / abs(true_derivative)

        # Print results
        print(f"{h:<12.2e} {fwd_error:<20.2e} {cent_error:<20.2e} {extra_error:<20.2e}")

        # Reduce step size
        h /= 2
````

Ok i think we can see that the error is bigger and bigger with decreasing h until 1e-16,
Now lets plot the `results` arrays we got earlier.

Because `results` is a dictionary and only handles a math function at a time lets make a function to plot its values, then call it twice
1 for cos, 1 for exp

````python
import matplotlib.pyplot as plt

def plot_errors(f, x_values, h_values):
    '''
    Plots log-log error vs step size
    '''
    for x in x_values:
        results = compute_errors(f, [x], h_values)[x]  #this is now a list, we’re evaluating the function at one or more x values.
        
        h_vals = np.array([res[0] for res in results])
        err_fwd = np.array([res[2] for res in results])
        err_cen = np.array([res[4] for res in results])
        err_ext = np.array([res[6] for res in results])

        plt.figure(figsize=(8, 6))
        plt.loglog(h_vals, err_fwd, 'r-', label='Forward Difference')
        plt.loglog(h_vals, err_cen, 'b--', label='Central Difference')
        plt.loglog(h_vals, err_ext, 'g-.', label='Extrapolated Difference')

        plt.xlabel(r'$\log_{10}h$')
        plt.ylabel(r'$\log_{10}|\varepsilon|$')
        plt.title(f'Error vs Step Size for {f.upper()} at x = {x}')
        plt.legend()
        plt.grid(True)
        plt.show()

# Plot for both functions
for func in ['cos', 'exp']:
    plot_errors(func, x_values, h_values)
````
