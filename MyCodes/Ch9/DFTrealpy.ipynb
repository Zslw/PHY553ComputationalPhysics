```python
""" From "COMPUTATIONAL PHYSICS: PROBLEM SOLVING with PYTHON"
    4th Ed. by RH Landau, MJ Paez, and CC Bordeianu (D)
    Copyright R Landau, Oregon State Unv, MJ Paez, Univ Antioquia, 
    C Bordeianu, Univ Bucharest, 2024. 
    Please respect copyright & acknowledge our work."""

# DFTreal.py:  Discrete Fourier Transform using real numbers

import numpy as np
import matplotlib.pyplot as plt
import math

N = 200                                   
Np = N 
signal = np.zeros((N+1), float)     
twopi = 2. * math.pi
sq2pi = 1. / math.sqrt(twopi) 
h = twopi / N
dftimag = np.zeros((Np), float)                              # Im. transform

t_values = []  # For plotting
signal_values = []

def f(signal):                                          
    step = twopi / N
    t = 0. 
    for i in range(0, N+1):
        signal[i] = 3 * math.sin(t * t * t)  
        t_values.append(t)
        signal_values.append(signal[i])
        t += step

def fourier(dftimag):                                     # DFT
    for n in range(0, Np):                                 
        imag = 0.                                         
        for k in range(0, N):                               
            imag += signal[k] * math.sin((twopi * k * n) / N) 
        dftimag[n] = -imag * sq2pi                           # Im transform

f(signal) 
fourier(dftimag)

# Plotting signal
plt.figure(figsize=(10, 4))
plt.plot(t_values, signal_values, color='orange')
plt.title('Signal y(t)= 3 sin(t)')
plt.xlabel('t')
plt.ylabel('signal')
plt.grid(True)
plt.show()

# Plotting imaginary part of DFT
plt.figure(figsize=(10, 4))
plt.bar(range(Np), dftimag, width=0.5, color='red')
plt.title('Fourier transform imaginary part')
plt.xlabel('Frequency index')
plt.ylabel('Transf.Imag')
plt.grid(True)
plt.show()
```
