```python
""" From "COMPUTATIONAL PHYSICS: PROBLEM SOLVING with PYTHON"
    4th Ed. by RH Landau, MJ Paez, and CC Bordeianu (D)
    Copyright R Landau, Oregon State Unv, MJ Paez, Univ Antioquia, 
    C Bordeianu, Univ Bucharest, 2024. 
    Please respect copyright & acknowledge our work."""

# NoiseSincFilter.py

import numpy as np
import matplotlib.pyplot as plt
import math
import random

max = 4000
array = np.zeros((max), float)
ps = np.zeros((max), float)
step = 2 * math.pi / 1000

t_vals = []
pure_signal = []
noisy_signal = []
filtered_signal = []

def function(array, max):
    f = np.zeros((max + 1), float) 
    step = 2 * math.pi / 1000
    x = 0. 
    for i in range(0, max):
        f[i] = 1 / (1. - 0.9 * math.sin(x))                             # Function
        array[i] = f[i] + 0.5 * (2. * random.random() - 1)              # Noise
        t_vals.append(x)
        pure_signal.append(f[i])
        noisy_signal.append(array[i])
        x += step

def filter():                           # Low-pass windowed sinc filter
    y = np.zeros((max), float)
    h = np.zeros((max), float)
    step = 2 * math.pi / 1000
    m = 100                                          # Set filter length
    fc = 0.07
    for i in range(0, 100):                          # Low-pass filter
        if ((i - (m / 2)) == 0):
            h[i] = 2 * math.pi * fc 
        else:
            h[i] = math.sin(2 * math.pi * fc * (i - m / 2)) / (i - m / 2)
        h[i] *= (0.54 - 0.46 * math.cos(2 * math.pi * i / m))  # Hamming window
    sum_h = sum(h[:100])                             # Normalize low-pass filter
    h[:100] /= sum_h
    for j in range(100, max - 1):                    # Convolute input + filter
        y[j] = 0.                    
        for i in range(0, 100):
            y[j] += array[j - i] * h[i]
    for j in range(0, max - 1):
        filtered_signal.append(y[j])

function(array, max)                                       
filter()     

# Plotting

plt.figure(figsize=(10, 4))
plt.plot(t_vals, pure_signal, color='red')
plt.title('Pure Signal')
plt.xlabel('t (s)')
plt.ylabel('f(t)')
plt.grid(True)
plt.xlim(0, 25)
plt.ylim(0, 10)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(t_vals, noisy_signal, color='red')
plt.title('Signal + Noise')
plt.xlabel('t (s)')
plt.ylabel('y(t)')
plt.grid(True)
plt.xlim(0, 25)
plt.ylim(0, 10)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(t_vals[:len(filtered_signal)], filtered_signal, color='red')
plt.title('Filtered Input')
plt.xlabel('t (s)')
plt.ylabel('y(t)')
plt.grid(True)
plt.xlim(0, 25)
plt.ylim(0, 10)
plt.show()
```
