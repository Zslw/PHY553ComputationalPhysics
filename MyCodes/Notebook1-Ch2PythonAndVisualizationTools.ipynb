**= does not mean equal**

Note that counter += 1 also is shorthand for counter = counter + 1. This second statement is a subtle but very important point. The = sign in a program is an *assignment* operator NOT an *equal* in the mathemetical sense. The code takes whatever is on the right side of the equal sign and replaces the variable on the left with that value. So counter = counter + 1, takes the old value of counter, adds one, and only then updates the variable counter with the new value.

````python
# PondMatPlot.py:    Monte-Carlo integration via vonNeumann rejection

import numpy as np
import matplotlib.pyplot as plt

N = 100;    Npts = 3000;    analyt = np.pi**2
x1 = np.arange(0, 2*np.pi+2*np.pi/N, 2*np.pi/N)
xi = [];    yi = [];    xo = [];    yo = []

fig, ax = plt.subplots()
y1 = x1 * np.sin(x1)**2                              # Integrand
ax.plot(x1,y1, 'c', linewidth=4)
ax.set_xlim( (0, 2*np.pi) )
ax.set_ylim( (0, 5) )
ax.set_xticks( [0, np.pi, 2*np.pi] )
ax.set_xticklabels( ['0', r'$\pi$', r'2$\pi$'] )
ax.set_ylabel(r'$ f(x) = x\, \sin^2 x $', fontsize=20)
ax.set_xlabel('x', fontsize=20)
fig.patch.set_visible(False)

def fx(x):
    return x*np.sin(x)**2                 # Integrand

j = 0                                               # Inside curve counter
xx = 2.*np.pi * np.random.rand(Npts)                # 0 =< x <= 2pi
yy = 5*np.random.rand(Npts)                         # 0 =< y <= 5

for i in range(1, Npts):
    if (yy[i] <= fx(xx[i])):
        if (i <= 100):
            xi.append(xx[i])
        if (i <= 100):
            yi.append(yy[i])
            j += 1
    else:
        if (i <= 100):
            yo.append(yy[i])
        if (i <= 100):
            xo.append(xx[i])
    boxarea = 2. * np.pi * 5.                      # Box area
    area = boxarea*j/(Npts-1)                     # Area under curve
    ax.plot(xo,yo,'bo',markersize=3)
    ax.plot(xi,yi,'ro',markersize=3)
    ax.set_title('Answers: Analytic = %5.3f, MC = %5.3f'%(analyt,area))

plt.show()
````

````python
from sympy import *

# Define symbols
x, y = symbols('x y')

# Derivative of tan(x)
y = diff(tan(x), x)
print("y =", y)  # Output: tan^2(x) + 1

# First derivative of 5 * x^4 + 7 * x^2 with respect to x
y = diff(5 * x**4 + 7 * x**2, x, 1)
print("y =", y)  # Output: 20*x^3 + 14*x
````

````python

# Series expansion of sin(x) around x = 0
sin_series_0 = sin(x).series(x, 0, 6)
print("sin(x) series around 0 =", sin_series_0)

# Series expansion of sin(x) around x = 10
sin_series_10 = sin(x).series(x, 10, 6)
print("sin(x) series around 10 =", sin_series_10)

# Series expansion of 1/cos(x) around x = 0
z = 1 / cos(x)
z_series = z.series(x, 0, 6)
print("1/cos(x) series around 0 =", z_series)

# Second derivative of 5 * x^4 + 7 * x^2 with respect to x
y = diff(5 * x**4 + 7 * x**2, x, 2)
print("Second derivative y =", y)  # Output: 60*x^2 + 14
````

````python
# Factorization examples
factor1 = factor(x**2 - 1)
print("Factor of x^2 - 1 =", factor1)  # Output: (x - 1)(x + 1)

factor2 = factor(x**3 - x**2 + x - 1)
print("Factor of x^3 - x^2 + x - 1 =", factor2)  # Output: (x - 1)(x^2 + 1)

# Simplification examples
simplify1 = simplify((x**3 + x**2 - x - 1) / (x**2 + 2 * x + 1))
print("Simplified expression 1 =", simplify1)  # Output: x - 1

simplify2 = simplify(x**3 + 3 * x**2 * y + 3 * x * y**2 + y**3)
print("Simplified expression 2 =", simplify2)  # Output: x^3 + 3*x^2*y + 3*x*y^2 + y^3

factor3 = factor(x**3 + 3 * x**2 * y + 3 * x * y**2 + y**3)
print("Factor of expression =", factor3)  # Output: (x + y)^3

# Trigonometric simplifications
simplify3 = simplify(1 + tan(x)**2)
print("Simplify 1 + tan^2(x) =", simplify3)  # Output: sec(x)^2

simplify4 = simplify(2 * tan(x) / (1 + tan(x)**2))
print("Simplify 2*tan(x)/(1+tan^2(x)) =", simplify4)  # Output: sin(2*x)
````
