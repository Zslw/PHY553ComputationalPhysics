```python
# BesselMatplotlib.py Spherical Bessel function by recursion
# From Landau's .../VPythonCodes/bessel.py website modified
# to use Matplotlib for graphics

import numpy as np
import matplotlib.pyplot as plt

Xmax = 40.
Xmin = 0.25
step = 0.1                       # Global class variable
order = 10; start = 50           # Plot j_order

def down(x, n, m):               # Method down, recurs downward
    j = np.zeros( (start + 2), float)
    j[m + 1] = j[m] = 1.         # Start with anything
    for k in range(m, 0, -1):
        j[k - 1] = ( (2.*k + 1.)/x)*j[k] - j[k + 1]
    scale = (np.sin(x)/x)/j[0]      # Scale solution to known j[0]
    return j[n]*scale

xAxis = []; yOrder = []; y1 = []
for x in np.arange(Xmin, Xmax, step):
    xAxis.append(x)
    yOrder.append( down(x, 0, start) )
    y1.append( down(x, 1, start) )

# Plot Bessel functions
plt.scatter(xAxis, yOrder)
plt.scatter(xAxis, y1)

# Print select Bessel function info (probably a better way than this.)
print('x        j3(x)          j5(x)          j8(x)')
print('0.1      ', down(0.1, 3, start), '  ', down(0.1, 5, start), '  ', down(0.1, 8, start) )
print('1      ',  down(1.0, 3, start), '  ', down(1.0, 5, start), '  ', down(1.0, 8, start) )
print('10      ', down(10., 3, start), '  ', down(10., 5, start), '  ', down(10., 8, start) )
```
